```python
import numpy as np
from math import *
import matplotlib.pyplot as plt



x = np.arange(0.000001,1,0.000001)


def f(x):
    return -(x*log2(x) + (1-x)*log2(1-x))

y = [f(i) for i in x]

plt.plot(x, y)
print(x[y.index(max(y))])
```
